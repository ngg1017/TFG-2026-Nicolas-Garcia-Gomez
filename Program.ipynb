```python
import os

os.environ["JAVA_HOME"] = "C:/Program Files/Java/jdk-21" 
os.environ["HADOOP_HOME"] = "C:/hadoop"

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName('primeros_pasos') \
    .master("local[*]") \
    .config("spark.driver.host", "localhost") \
    .getOrCreate()

sc = spark.sparkContext
from pyspark.sql.types import *

print("¡Si ves esto, Spark ya está funcionando!")
```

```python
df_principal = spark.read.format ("csv").option("header", "true").load("Practica_Provisionales.csv").persist()

df_principal.show(1)
```

```python
import os
import sys

os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
from pyspark.sql.functions import split, date_sub, to_date

a = df_principal.select(
    date_sub(
        to_date(split(df_principal["Hora alta"], " ")[0], "dd/MM/yyyy"), 
        3
    ).alias("fecha_ajustada")
)

a.show(5)

# Convertir DataFrame a RDD
rdd_principal = df_principal.rdd

# Ver los primeros 5 elementos
print(rdd_principal.take(5))
```
